```
!pip install --upgrade covid19umbria

import datetime
import matplotlib.pyplot as plt
from covid19umbria import Covid19Umbria

# how many days
NUMDAYS = 30

base = datetime.datetime.today()
date_list = [base - datetime.timedelta(days=x) for x in range(NUMDAYS)]

covid19 = Covid19Umbria()

# fetch data
x = date_list
y = [covid19.get_current_active_cases(date) for date in date_list]
plt.plot(x, y, label="Current Active Cases")

# plot graph
plt.plot()
plt.title("Covid19Umbria")
plt.legend()
plt.show()
```
